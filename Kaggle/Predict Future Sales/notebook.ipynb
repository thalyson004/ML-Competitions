```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns 
from googletrans import Translator

from sklearn.model_selection import train_test_split

%matplotlib inline
mpl.rcParams["figure.dpi"] = 300
```

```python
# Read data
df_items_category = pd.read_csv("data/item_categories.csv")
df_items = pd.read_csv("data/items.csv")
df_shops = pd.read_csv("data/shops.csv")

df_sales_train = pd.read_csv("data/sales_train.csv")

df_sample_submission = pd.read_csv("data/sample_submission.csv")
df_test = pd.read_csv("data/test.csv")
```

```python
df_small =  train_test_split(df_sales_train, train_size= 0.95)[1]
```

```python
df_small.columns
```

```python
df_small.sort_values(by='date_block_num')
```

```python
sns.catplot(data=df_small, x="date_block_num", y="item_cnt_day", height=5, aspect=2)
```

```python
g = sns.catplot(
    data=df_small.groupby('date_block_num', as_index=False).sum(), 
    x='date_block_num', 
    y='item_cnt_day',
    kind="bar",
    height=5,
    aspect=2,
)
g.set_axis_labels("Months", "Sold items")
plt.title('Sold items by month')
```

```python
sns.catplot(
    data=df_small.groupby('date_block_num', as_index=False).sum(), 
    x='date_block_num', 
    y='item_price',
    kind="bar",
    height=5,
    aspect=2
)
```
